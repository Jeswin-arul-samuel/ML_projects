```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("salaries_dataset.csv", sep=';')
df.head()
```

```python
df.shape
```

```python
plt.scatter('experience_level','salary', data =df)
plt.xlabel('Exp_level')
plt.ylabel('Salary')
plt.title('Exp_level vs Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()
linreg.fit(df[['experience_level']],df['salary'])
```

```python
plt.scatter('experience_level','salary', data =df)
plt.plot(df['experience_level'], linreg.predict(df[['experience_level']]), c='g', label = 'linear model')
plt.xlabel('Exp_level')
plt.ylabel('Salary')
plt.legend()
plt.title('Exp_level vs Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

polyreg = PolynomialFeatures(degree=4)
exp_level_poly = polyreg.fit_transform(df[['experience_level']])
```

```python
exp_level_poly
```

```python
lin_pol_reg = LinearRegression()
lin_pol_reg.fit(exp_level_poly, df['salary'])
```

```python
plt.scatter('experience_level','salary', data =df)
plt.plot(df['experience_level'], lin_pol_reg.predict(exp_level_poly), c='g', label = 'poly model')
plt.xlabel('Exp_level')
plt.ylabel('Salary')
plt.legend()
plt.title('Exp_level vs Salary')
plt.show()
```

```python
new_employee_exp = float(input("Enter the employees years of experience:"))
```

```python
new_exp = polyreg.transform([[new_employee_exp]])
new_exp
```

```python
print("The new employee salary is {}" .format(lin_pol_reg.predict(new_exp)))
```
